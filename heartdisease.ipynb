```python
import numpy as np
import pandas as pd
from sklearn.model_selection  import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/Heart_Disease_Prediction.csv')
data.head(6)
```

```python
data.shape
```

```python
x= data.drop(columns = 'Heart Disease', axis = 1)
y = data['Heart Disease']
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.1, stratify = y, random_state = 2
                                                    )
```

```python
x.shape, x_train.shape
```

```python
y.shape, y_train.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_trainpred = model.predict(x_train)
trainingaccuracy = accuracy_score(x_trainpred, y_train)
trainingaccuracy
```

```python
x_testpred = model.predict(x_test)
testaccuracy = accuracy_score(x_testpred, y_test)
testaccuracy
```
